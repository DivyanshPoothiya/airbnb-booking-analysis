# Handling Missing Values

```python
print(df.fillna({'reviews_per_month' : 0, 'review_rate_number' : df['review_rate_number'].min()},inplace=True))
```

```python
df.dropna(subset=['name','host_name'],inplace = True)
```

```python
df.isnull().sum()
```

```python
df.dropna(subset=['neighbourhood_group','neighbourhood'],inplace = True)
```

```python
df.isnull().sum()
```

```python
print(df.dropna(subset=['house_rules','license'],inplace=True))
```

```python
print(df.dropna(subset=['price'],inplace=True))
```

```python
df.isnull().sum()
```

```python
df=df.drop(columns=['license','house_rules'],errors='ignore')
```

# Visualization 

```python
plt.figure(figsize=(4,4))
sns.histplot(x='price',data=df,bins=10,kde=True) 
plt.title('Distribution of listing price')
plt.xlabel('Price $',rotation=0)
plt.ylabel('Frequency')
plt.show()           
```

```python
df['room_type']
```

# Distributed Room Type

```python
plt.figure(figsize=(5,5))
sns.countplot(x='room_type',data=df,color='green')
plt.title("Room Type")
plt.xlabel("Rooms")
plt.ylabel("count")
plt.show()
```

# Different Neighbourhood 

```python
plt.figure(figsize=(5,4))
sns.countplot(y='neighbourhood_group',data=df,order=df['neighbourhood_group'].value_counts().index)
plt.title("Neighbourhood")
plt.ylabel("Neighbour lisitings in groups")
plt.xlabel("count")
plt.show()
```

# Relation Between Price and Room Type

```python
plt.figure(figsize=(10,6))
sns.boxplot(x ='room_type',y ='price',hue='price',data=df,palette ='Set1')
plt.title('Price vs Room type')
plt.xlabel("Room Type")
plt.ylabel("Price ($)")
plt.legend(title='Room Type')
plt.show()
```
